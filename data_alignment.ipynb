```python
import pandas as pd
import os
# Correct path to the file
file_path = 'C:/Users/ASUS/Desktop/Msc Course/CT calibration/Dataset/SNR_CNR/Khana.xlsx'

# Load the Excel file with all sheets
data = pd.read_excel(file_path, sheet_name=None)

# Dictionary to hold data for all sheets processed
processed_data_corrected = {}

# Process each sheet similarly
for sheet_name, sheet_data in data.items():
    # Initialize lists to store the resulting data
    img_nums = []
    stack_nos = []
    snr_values = []
    cnr_values = []

    # Iterate through each block of 13 rows and each of the 7 columns for SNR and CNR
    for block in range(25):  # 25 images
        for col in range(7):  # 7 columns for SNR (columns 0 to 6)
            for row in range(13):  # 13 values per image
                img_nums.append(block + 1)  # Image number
                stack_nos.append(f"Stack{col * 13 + row + 1}")  # Stack number, reset each image
                if (block * 13 + row) < len(sheet_data):
                    snr_index = col  # SNR values from the first 7 columns
                    cnr_index = col + 8  # CNR values from columns I to O (index 8 to 14)
                    snr_values.append(sheet_data.iloc[block * 13 + row, snr_index])  # SNR value
                    cnr_values.append(sheet_data.iloc[block * 13 + row, cnr_index])  # CNR value
                else:
                    snr_values.append(None)  # Append None if out of bounds
                    cnr_values.append(None)

    # Create a new DataFrame with the structured data
    processed_sheet = pd.DataFrame({
        "ImgNum": img_nums,
        "StackNo": stack_nos,
        "SNR": snr_values,
        "CNR": cnr_values
    })

    # Add to the dictionary of processed data
    processed_data_corrected[sheet_name] = processed_sheet

# Save the new DataFrame to a new Excel file with sheet names retained
file_path = 'C:/Users/ASUS/Desktop/Msc Course/CT calibration/Dataset/SNR_CNR/Khana_processed.xlsx'

# Check if the file exists
if not os.path.exists(file_path):
    # Create a new empty DataFrame
    df = pd.DataFrame()
    
    # Save the empty DataFrame to the specified path
    df.to_excel(file_path, index=False)
    print(f"File created at {file_path}")
else:
    print(f"File already exists at {file_path}")
output_path_all_sheets_corrected = 'C:/Users/ASUS/Desktop/Msc Course/CT calibration/Dataset/SNR_CNR/Khana_processed.xlsx'
with pd.ExcelWriter(output_path_all_sheets_corrected, mode='a', engine='openpyxl', if_sheet_exists='replace') as writer:  # Open in append mode using openpyxl
    for sheet_name, sheet_data in processed_data_corrected.items():
        sheet_data.to_excel(writer, sheet_name=sheet_name, index=False)

print("Data processed and appended to:", output_path_all_sheets_corrected)
```
